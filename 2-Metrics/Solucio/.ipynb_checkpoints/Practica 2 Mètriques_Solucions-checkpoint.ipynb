```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn import datasets
from sklearn.model_selection import train_test_split , KFold
from sklearn.preprocessing import Normalizer
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, roc_curve, roc_auc_score
from sklearn.neighbors import KNeighborsClassifier

from collections import Counter
```

```python
# import iris dataset
iris = datasets.load_iris()
# np.c_ is the numpy concatenate function
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                      columns= iris['feature_names'] + ['target'])
```

```python
def Normalize(x_train) -> object:
    scale = Normalizer().fit(x_train) # the scaler is fitted to the training set
    return scale

def Knn_train_and_predict(x_train,y_train, x_test,K=3) -> list:

    knn=KNeighborsClassifier(K)
    knn.fit(x_train, y_train)
    print("Trained")
    y_pred_sklearn= knn.predict(x_test)
    return y_pred_sklearn, knn



```

De totes maneres hi ha més maneres d'evaluar com és la performance d'un model que es relacionen amb la precissió i el recall.
En aquest cas dos gràfiques que ajuden a medir com de bo és el model són la **ROC curve i la PR curve**.

<img src="images/Roc_curve.svg.png" width="300">


La corba ROC (receiver operating characteristic) relaciona el recall amb el ràtio de falsos positius. És a dir relaciona la sensibilitat del nostre model amb les fallades optimistes (classificar els negatius com a positius). Té sentit ja que, generalment, si augmentem el recall, el nostre model tendirà a ser més optimista i introduirà mes falsos positius en la classificació.

En les corbes ROC, ens interessa que la corba s'acosti el màxim possible a la cantonada superior esquerra de la gràfica, de manera que el fet d'augmentar la sensibilitat (el recall) no faci que el nostre model introdueixi més falsos positius.


<img src="images/PR_curve.png" width="300">



La corba PR és el resultat de dibuixar la gràfica entre el precision i el recall. Aquesta gràfica ens permet veure a partir de què recall tenim una degradació de la precisió i viceversa. L'ideal seria una corba que s'acosti el màxim possible a la cantonada superior dreta (alta precisió i alt recall).

La manera d'evaluar un PR normalment s'anomena MAP (mean average precission) i és una manera de calcular l'àrea sota la corba PR o PR AUC, o cosa que és el mateix, el resultat d'integrar la corba. El Average Precision ens serveix per a avaluar i comparar el rendiment de models. Com més s'acosti el seu valor a 1, millor serà el nostre model.

### Hi ha maneres de calcular el millor threshold un exemple és la G-mean i la J-statistic

<img src="images/Thres.png" width="1000">


```python
def J_statistic(y_true:list, y_score:list):
    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    # get the best threshold
    J = tpr - fpr
    ix = np.argmax(J)
    best_thresh = thresholds[ix]
    print('Best Threshold=%f' % (best_thresh))
    
        # plot the roc curve for the model
    plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
    plt.plot(fpr, tpr, marker='.', label='Logistic')
    plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label='Best')
    # axis labels
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    # show the plot
    plt.show()
    return best_thresh
    

def G_mean(y_true:list, y_score:list):
    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    # calculate the g-mean for each threshold
    gmeans = np.sqrt(tpr * (1-fpr))
    # locate the index of the largest g-mean
    ix = np.argmax(gmeans)
    print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))
    # plot the roc curve for the model
    plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
    plt.plot(fpr, tpr, marker='.', label='Logistic')
    plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label='Best')
    # axis labels
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    # show the plot
    plt.show()
    return thresholds[ix]
```

### Anem a implementar un K-Fold !! 


La validació creuada o cross-validation és una tècnica utilitzada per a avaluar els resultats d'una anàlisi estadística i garantir que són independents de la partició entre dades d'entrenament i prova. Consisteix a repetir i calcular la mitjana aritmètica obtinguda de les mesures d'avaluació sobre diferents particions.

El mètode K-Fold Cross-Validation és també un procés iteratiu. Consisteix a dividir les dades de manera aleatòria en k grups d'aproximadament la mateixa grandària, k-1 grups s'empren per a entrenar el model i un dels grups s'empra com a validació. Aquest procés es repeteix k vegades utilitzant un grup diferent com a validació en cada iteració. El procés genera k estimacions de l'error la mitjana del qual s'empra com a estimació final.

Dos avantatges del mètode K-Fold Cross-Validation enfront del LOOCV:

Requeriments computacionals: el nombre d'iteracions necessàries ve determinat pel valor k triat. En general, es recomana un k entre 5 i 10.

Balanç entre bias i variància: el principal avantatge de K-fold CV és que aconsegueix una estimació precisa de l'error de test gràcies a un millor balanç entre bias i variància. LOOCV empra n-1 observacions per a entrenar el model, la qual cosa és pràcticament tot el set de dades disponible, maximitzant així l'ajust del model a les dades disponibles i reduint el bias. No obstant això, per a l'estimació final de l'error es fan una mitjana de les estimacions de n models entrenats amb pràcticament les mateixes dades (només hi ha una dada de diferència entre cada conjunt d'entrenament), per la qual cosa estan altament correlacionats. Això es tradueix en un major risc de overfitting i per tant de variància. En el mètode K-fold CV els k grups emprats com a entrenament són molt menys solapants, la qual cosa es tradueix en menor variància en fer una mitjana de les estimacions d'error.



<img src="images/Cross.png" width="1000">

```python
def Customed_Kfold(X, y, n_splits):
    kf = KFold(n_splits=n_splits)
    k_values = np.arange(1,30,2)
    acc_k = []
    
    for k in k_values:
        accuracy_by_fold = 0
        knn = KNeighborsClassifier(k)
        
        for x_train_fold_idx, x_valid_fold_idx in kf.split(X):
            
            ## The train partition i (n_splits-1)
            x_train_fold = X[x_train_fold_idx]
            y_train_fold = y[x_train_fold_idx]
            
            ## The test partition (else)
            x_test_fold = X[x_valid_fold_idx]
            y_test_fold = y[x_valid_fold_idx]
            
            ## Fit the model
            knn.fit(x_train_fold, y_train_fold)
            y_pred_fold = knn.predict(x_test_fold)
            
            accuracy_by_fold += accuracy_score(y_pred_fold, y_test_fold) ## Accumulates the accuracyç
        
        accuracy_fold= accuracy_by_fold/ n_splits ## Divide by the number of splits
        acc_k.append(accuracy_fold)
    
    return acc_k, k_values
            
```

```python
accuracies, ks = Customed_Kfold(X, y, 10) # Pima
```

```python
print(f'The accuracy for each K value was {list ( zip (accuracies, ks))}')
print(f'Best accuracy was {np.max(accuracies)}, which corresponds to a value of K= {ks[np.argmax(accuracies)]}')
```

## Resposta Exercici 1

**Normalitzar** reescala el vector perquè cada mostra tingui una norma unitària, independentment de la distribució de les mostres. i algunes característiques originals tinguessin una combinació de valors positius i negatius.

**StandardScaler** elimina la mitjana i escala les dades a la variància de la unitat. L'escalat redueix l'interval dels valors de les característiques tal com es mostra a la figura esquerra de sota. No obstant això, els valors atípics tenen una influència en calcular la mitjana empírica i la desviació estàndard. Cal tenir en compte, en particular, que com que els valors atípics de cada característica tenen diferents magnituds, la propagació de les dades transformades en cada característica és molt diferent: la majoria de les dades es troben en l'interval [-2, 4] per a la funció d'ingressos mitjans transformats, mentre que les mateixes dades es retallen en el rang més petit [-0,2, 0.2] per a l'ocupació mitjana de la casa transformada. Per tant el Standard Scaler no pot garantir escales de característiques equilibrades en presència d'outliers


**MinMaxScaler** reescala el conjunt de dades de manera que tots els valors de les característiques estiguin a l'interval [0, 1] com es mostra al plafó dret de sota. No obstant això, aquest escalat comprimeix tots els inliers en un rang estret [0, 0.005] per a l'ocupació mitjana de la casa transformada.

**L'escalador estàndard i escalador de mínims són molt sensibles a la presència d'outliers.**


**MaxAbsScaler** és similar a MinMaxScaler, excepte que els valors s'assignen a través de diversos intervals depenent de si hi ha valors negatius o positius. Si només hi ha valors positius, l'interval és [0, 1]. Si només hi ha valors negatius, el rang és [-1, 0]. Si hi ha valors negatius i positius, el rang és [-1, 1]. En dades només positives, tant MinMaxScaler com MaxAbsScaler es comporten de manera similar. Per tant, MaxAbsScaler també pateix de la presència de grans outliers. MaxAbsScaler va ser dissenyat específicament per a escalar dades disperses


A diferència dels escaladors anteriors, **RobustScaler** es basa en percentils i, per tant, no està influenciada per un petit nombre d'outliers molt grans. En conseqüència, l'interval resultant dels valors de les característiques transformades és més gran que per als escaladors anteriors i, més important, són aproximadament similars: per a ambdues característiques la majoria dels valors transformats es troben en un rang [-2, 3]. Tingueu en compte que els outliers encara estan presents en les dades transformades. El RobustScaler no es pot ajustar a entrades disperses. Si les vostres dades contenen molts valors atípics, és probable que l'escalat utilitzant la mitjana i la variància de les dades no funcioni molt bé. En aquests casos, podeu utilitzar RobustScaler com a substitució desplegable. Utilitza estimacions més robustes per al centre i l'interval de les vostres dades.

## Escalats no lineas

Quan parlem d'escalats no lineals us eu d'imaginar una aplicació no lineal entre dos espais vectorials i això passa quan no es compleix (en essencia) això:

Donats dos espais vectorials V i W, i donada una aplicació f: V W, direm que f és
lineal si conserva les combinacions lineals, és a dir: donada una combinació lineal entre
vectors de V, les seves imatges en W verifiquen la mateixa combinació:
→
si o = α v+ w (en V) llavors o’ = α v’ + w’ (en W)β β
on o’, v’, w’ són respectivament les imatges d'o, v, w.



**PowerTransformer** són una família de transformacions paramètriques i monotòniques que s'apliquen per fer les dades més similars a gaussianes. Això és útil per modelar problemes relacionats amb l'heteroscedasticitat (variància no constant), o altres situacions on es desitja la normalitat.

**QuantileTransformer** aplica una transformació no lineal de tal manera que la funció de densitat de probabilitat de cada característica es maparà a una distribució uniforme o gaussiana. En aquest cas, totes les dades, incloent valors atípics, s'assignaran a una distribució uniforme amb l'interval [0, 1], fent que els valors atípics siguin indistingibles dels valors atípics.

**RobustScaler i QuantileTransformer** són robustos per a valors atípics en el sentit que afegir o eliminar valors atípics en el conjunt d'entrenament produirà aproximadament la mateixa transformació. Però contràriament a RobustScaler, QuantileTransformer també col·lapsarà automàticament qualsevol outlier establint-los als límits de rang definits a priori (0 i 1). Això pot resultar en artefactes de saturació per a outliers.

```python
class Confusion_Matrix(object):
    
    
    def __init__(self, y_pred :list, gt:list):
        assert len(y_pred) == len(gt), "Les prediccions i el gt tenen shapes diferents"
        self._y_pred = y_pred
        self._gt = gt
        
        self._tp, self._fp, self._tn, self._fn = self.get_confusion()
        
        
    def get_confusion(self):
        tp = fp = tn = fn = 0
        
        compare = zip(self._y_pred, self._gt)
        
        for pred, gt in compare:
            if pred==gt==1:
                tp+=1
            elif (pred==1) and (gt!=pred):
                fp +=1
            elif (pred==0) and (gt!=pred):
                fn+=1
            else:
                tn+=1
        
        return tp, fp, tn, fn
    
    def precision(self):
        return self._tp/(self._tp + self._fp)
    
    def recall(self):
        return self._tp/(self._tp+self._fn)
    
    def specificity(self):
        return self._tn/(self._tn + self._fp)
    
    def f_score(self):
        return (2*self.precision() * self.recall())/(self.precision() +  self.recall())
            
```

```python
cm = Confusion_Matrix(y_pred=[1,1,1,1,1,1,1,1,0],gt=[1,1,1,0,0,0,0,1,0])
```

```python
cm.precision() # En aquest la precisio es del 50% per que dels 8 valors predits com a classe 1 al gt hi ha 4 per tant l'encert dels valors que són 1 és del 50%
```

```python
cm.recall() # En contraposició a la precissió com que totes les classe spositives han estat ven predites la sensivilitat del model a la classe positiva és del 100%
```

```python
cm.specificity() # Respecte ala
```

```python
cm.f_score()
```

## Exercici 4

En aquest cas com que estic treballant sobre dades mèdiques i assumeixo que no puc optimitzar el model de manera òptima prioritzo el recall.

```python
def Customed_Kfold(X, y, n_splits):
    kf = KFold(n_splits=n_splits)
    k_values = np.arange(1,30,2)
    acc_k = []
    
    for k in k_values:
        recall_by_fold = 0
        knn = KNeighborsClassifier(k)
        
        for x_train_fold_idx, x_valid_fold_idx in kf.split(X):
            
            ## The train partition i (n_splits-1)
            x_train_fold = X[x_train_fold_idx]
            y_train_fold = y[x_train_fold_idx]
            
            ## The test partition (else)
            x_test_fold = X[x_valid_fold_idx]
            y_test_fold = y[x_valid_fold_idx]
            
            ## Fit the model
            knn.fit(x_train_fold, y_train_fold)
            y_pred_fold = knn.predict(x_test_fold)
            
            cm = Confusion_Matrix(y_pred_fold, y_test_fold)
            
            recall_by_fold += cm.recall()
        
        recall_fold = recall_by_fold/ n_splits ## Divide by the number of splits
        acc_k.append(recall_fold)
    
    return acc_k, k_values
            
```

```python
recalls, ks = Customed_Kfold(X, y, 10) # Pima
```

```python
print(f'The recall for each K value was {list ( zip (recalls, ks))}')
print(f'Best recall was {np.max(recalls)}, which corresponds to a value of K= {ks[np.argmax(recalls)]}')
```
